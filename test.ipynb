```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
gamma = 60/100
```

```python
table = (np.arange(0, 256.0 / 255, 1.0 / 255) ** gamma) * 255
```

```python
oritab = ((np.arange(0, 256.0 / 255, 1.0 / 255) ** 1) * 255).astype(int)
oritab = np.reshape(oritab,(16,16))
```

```python
#ACC
import matplotlib
my_cmap = matplotlib.cm.get_cmap('rainbow')
my_cmap.set_under('w')

fig = plt.figure(figsize=(12.5,12.5))

ax = fig.add_subplot(111)
ax.grid(True, which='minor', axis='both', linestyle='-', color='k')

heatmap = ax.pcolor(oritab, cmap=my_cmap,facecolor='black', edgecolor='black')

# ax.set_xticklabels(np.arange(16), rotation=0, fontsize=22)
# ax.set_yticklabels(np.arange(16), rotation=0, fontsize=22)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
# ax.yaxis.tick_left(False)   

for a in range(oritab.shape[0]):
    for b in range(oritab.shape[1]):
        if oritab[a, b] <= 20:
            plt.text(b + 0.5, a + 0.5,  oritab[a, b],
                 horizontalalignment='center',
                 verticalalignment='center',
                 color = 'white',size=22
                 )
        elif oritab[a, b] <= 90 and oritab[a, b] > 33:
            plt.text(b + 0.5, a + 0.5,  oritab[a, b],
                 horizontalalignment='center',
                 verticalalignment='center',
                 color = 'black',size=22
                 )
        elif oritab[a, b] > 230:
            plt.text(b + 0.5, a + 0.5,  oritab[a, b],
                 horizontalalignment='center',
                 verticalalignment='center',
                 color = 'yellow',size=22
                 )
        else:
            plt.text(b + 0.5, a + 0.5,  oritab[a, b],
                 horizontalalignment='center',
                 verticalalignment='center',size=22
                 )
plt.show()
```

```python
img = (ori[0]*255).astype(np.uint8)
width,height = img.shape[:2]
dx=0
dy=0
k=-0.5
```

```python


fx = width
fy = height

cx = width * 0.5 + dx
cy = height * 0.5 + dy

camera_matrix = np.array([[fx, 0, cx], [0, fy, cy], [0, 0, 1]], dtype=np.float32)
```

```python
camera_matrix
```

```python
distortion = np.array([k, k, 0, 0, 0], dtype=np.float32)
map1, map2 = cv2.initUndistortRectifyMap(camera_matrix, distortion, None, None, (width, height), cv2.CV_32FC1)
img = cv2.remap(img, map1, map2, interpolation=cv2.INTER_LINEAR, borderMode=0, borderValue=0)
```
